```python
%matplotlib inline
import pickle, os, tqdm, json
import numpy as np
import matplotlib.pyplot as plt

fontsize = 30
label_fontsize = 40
linewidth = 5
import matplotlib.font_manager as font_manager
fname='/data/shaoerzhuo/Fonts/'
font_dirs = [fname, ]
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
font_list = font_manager.createFontList(font_files)
font_manager.fontManager.ttflist.extend(font_list)
plt.rcParams["font.family"] = "Times New Roman"
plt.rc('font', family='Times New Roman', size=fontsize)
plot_dict = {'markersize':15, 'markeredgewidth':5, 'linewidth':linewidth}
c = [[0, 0.4470, 0.7410], [0.8500, 0.3250, 0.0980], [0.9290, 0.6940, 0.1250],[0.494, 0.184, 0.556], [0.466, 0.674, 0.188], [0.301, 0.745, 0.933], [0.635, 0.078, 0.184]]

fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111)

ax.plot(range(0, 20), np.load('PLANBERT.npy', allow_pickle=True).tolist()[0][3][1], color=c[0], marker='x', linestyle='-', **plot_dict)
ax.plot(range(3, 20), np.load('PLANBERT.npy', allow_pickle=True).tolist()[3][3][1], color=c[1], marker='o', linestyle='-', **plot_dict)
ax.plot(range(5, 20), np.load('PLANBERT.npy', allow_pickle=True).tolist()[5][3][1], color=c[2], marker='+', linestyle='-', **plot_dict)
ax.plot(range(10, 20), np.load('PLANBERT.npy', allow_pickle=True).tolist()[10][3][1], color=c[3], marker='h', linestyle='-', **plot_dict)

ax.xaxis.set_ticks(np.arange(20))

ax.legend([
    'h=0', 'h=3', 'h=5', 'h=10', 
], fontsize=30,  ncol=2, loc='best')

plt.ylabel('Recall@10 for (N-60)', fontsize=label_fontsize)
plt.xlabel('Reference Items', fontsize=label_fontsize)
plt.title('Recall@10 per Time Slots')
```
